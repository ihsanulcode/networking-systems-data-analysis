## Import libraries

```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load the Excel file
df = pd.read_excel('Project Survey Report.xlsx')
df.head()
```

## Cleaning the Dataset

### Dropping columns that are necessary for our task

```python
# List of columns to drop
columns_to_drop = [0, 1, 16]

# Drop the specified columns
df.drop(df.columns[columns_to_drop], axis=1, inplace=True)
df.head()
```

### Rename columns

```python
# Dictionary to map column indices to new names
columns_rename = {
    0: 'gender',
    1: 'country',
    2: 'profession',
    3: 'age',
    4: 'ever_think_about_cybsec_space-based_netw_sys',
    5: 'ever_discussed_about_netw_sys',
    6: 'concerned_about_dangerous_impact_secure_data',
    7: 'willing_to_switch_to_secured_netw_2035',
    8: 'main_concern_regarding_transition_to_5G_2035',
    9: 'feel_about_the_global_plan_to_adapt_secured_netw_sys_2035',
    10: 'blv_industries_are_prepared_secured_netw_sys_2035',
    11: 'perceive_govt_efforts_in_secured_netw_sys',
    12: 'transition_to_having_5g_network_will_impact_dailylife',
    13: 'prefer_to_use_secured_5g'
}

# Rename the columns
df.rename(columns={df.columns[idx]: new_name for idx, new_name in columns_rename.items()}, inplace=True)
df.info()
```
