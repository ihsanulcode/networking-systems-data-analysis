#### Handle Missing Values

```python
# Count missing values in each column
missing_values = df.isnull().sum()
missing_values
```

```python
# Handle emails column
# Function to check for "@" in the email and replace invalid ones with "No"
def check_and_replace_email(email):
    if pd.isna(email) or "@" not in email:
        return "No"
    return "Yes"

# Apply the function to the 'emails' column
df['happy_to_provide_emails'] = df['happy_to_provide_emails'].apply(check_and_replace_email)
df['happy_to_provide_emails'].value_counts()
```

```python
# Count missing values in each column
missing_values = df.isnull().sum()
missing_values
```

```python
# Loop through all columns
for column in df.columns:
    if df[column].dtype == 'object':  # Check if the column is categorical
        mode_value = df[column].mode()[0]  # Calculate the mode
        df[column].fillna(mode_value, inplace=True)  # Fill missing values with the mode
```

```python
# Count missing values in each column
missing_values = df.isnull().sum()
missing_values
```

#### Label encoding
using label encoder

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
lst_encoders = []
lst_labels = []
for col in df.columns:
    encoding = encoder.fit_transform(df[col])
    lst_encoders.append(encoder.classes_)
    lst_labels.append(encoder.transform(encoder.classes_))
    # Print the encoding values
    print(col)
    print("Labels: ",encoder.classes_)
    print("Encodings:",encoder.transform(encoder.classes_))
    print('----------------------------------------------')
    # Replace encoded lebsl with main df
    df[col] = encoding
```

### Generate synthetic data

```python
df.info()
```

```python
import pandas as pd
from sklearn.utils import resample

# Function to generate synthetic data using resampling (data augmentation)
def generate_synthetic_data(data, num_samples=1000):
    synthetic_data = pd.DataFrame(columns=data.columns)
    for category in data['prefer_to_use_secured_5g'].unique():
        category_data = data[data['prefer_to_use_secured_5g'] == category]
        synthetic_samples = resample(category_data, replace=True, n_samples=num_samples, random_state=42)
        synthetic_data = pd.concat([synthetic_data, synthetic_samples])
    return synthetic_data

# Generate synthetic data
synthetic_data = generate_synthetic_data(df)
synthetic_data.info()
```
